```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,180)
y = np.sin(x * np.pi / 180)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)
plt.plot(x,y)
plt.xlim(-30,390)
plt.ylim(-1.5,1.5)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("The Title")
plt.show()
```
